```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([14,24,32,43,51,68,73,82,87,98])
y = np.array([5,16,28,34,45,54,68,76,85,96])
z = np.array([16,27,33,42,56,64,78,87,91,97])
```

```python
# Line Plot
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker = 'o' , mfc = 'red', color= 'black' , ls= ':', ms=5 , label= '2024');
plt.grid()
plt.legend()
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x,z,marker = 'v' , mfc = 'blue', color= 'black' , ls= '--', label='2019');
plt.plot(x,y,marker = 'o' , mfc = 'red', color= 'black' , ls= ':', ms=5, label='2025');

plt.grid()
plt.legend()
```

```python
# subplot
plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x,y,marker = 'o' , mfc = 'red', color= 'black' , ls= ':' );
plt.grid()

plt.subplot(2,2,3)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x,z,marker = 'v' , mfc = 'blue', color= 'black' , ls= '--');
plt.grid()
```

```python
# Bar Plot
```

```python
courses = ['java', 'python','cpp','c','R']
students = [67,63,59,55,47]
```

```python
plt.bar(courses,students,color='black', edgecolor='red',width=0.8, linewidth=3,alpha=0.7)
plt.title('Bar Plot')
plt.xlabel('Courses')
plt.ylabel('Students')
plt.grid()
```
